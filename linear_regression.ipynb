using a Data set from kaggle.com  <br>

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("placement.csv")
```

```python
df
```

# Removing unwanted Data

```python
df.drop(["sl_no","ssc_b","hsc_s" , "hsc_b", "specialisation","mba_p","status","etest_p"], axis=1 , inplace=True)
df
```

```python
df.rename(columns={"ssc_p" : "10th_marks" , "hsc_p": "12th_marks" ,
  "degree_p" : "UG_Score" , "degree_t" : "UG_Degree" 
  , "workex":"work_experience" }, inplace=True)
```

```python
df
```

```python
df.UG_Degree.unique()
```

# Changing in work experience : <br> 
##  yes  = 1 <br>
##  no = 0 <br>

```python
def gen(x):
    if x == 'No':
        return 0 
    elif x == 'Yes':
        return 1           

df.work_experience  =  df.work_experience.apply(gen)

df 
```

```python
df.salary = df.salary.fillna(0)
df
```

# Changing in Gender : <br>
##  we use one hot encoding <br>
  <br>
  
# Changing in UG_Degree <br>

##  Sci&Tech  = 0 <br>
## using one hot encoding
  <br>

```python
dum1 =pd.get_dummies(df.gender)
dum2 = pd.get_dummies(df.UG_Degree)
merged = pd.concat([df,dum1,dum2],axis='columns')
merged
```

# Now we will remove gender and UG_degree columns as we don't need it.

```python
merged.drop(["gender","F","UG_Degree","Others"], axis=1 , inplace=True)
merged.rename(columns={"Sci&Tech" : "sci", "Comm&Mgmt" : "comm"} ,inplace=True)
merged
```

# Finally we have Cleaned Data

```python
X = merged.drop(["salary"], axis=1)
y = merged.salary
y
```

# Finally saving our datafram as acsv file 

```python
merged.to_csv('final_data.csv')
```

```python
merged
```

# Now we will apply Linear regration to train our model

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression( )
```

```python
from sklearn.model_selection import train_test_split 
x_train , x_test , y_train , y_test = train_test_split(X,y , train_size=0.1)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
```
